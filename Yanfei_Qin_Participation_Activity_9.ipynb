1. Right now, the data is somewhere between long and wide format. Use melt() to make the data completely long format. Hint: data and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, close, and volume. Print the first 5 rows of the dataframe in long format. (3 points)

2. Calculate the rolling 60-day aggregations of the OHLC (open, high, low, close)  and volume data by ticker for the FAANG data. Make the following aggregations:  (5 points)

A. Mean of the opening price 

B. Maximum of the high price

C. Minimum of the low price

D. Mean of the closing prince 

E. Sum of the volume traded

```python
faang = pd.read_csv('faang.csv', index_col='date', parse_dates=True)
faang
```

```python
import numpy as np

faang.groupby('ticker').rolling('60D').agg(
    {
        'open': np.mean, #A
        'high': np.max, #B
        'low': np.min, #C
        'close': np.mean, #D
        'volume': np.sum #E
    }
)
```

4. Calculate the Z-scores for each numeric column of Amazon's data (ticker os AMZN) in Q4 2018 using apply().(3 points)


```python
AMZN_z_scores = faang.loc['2018-Q4'].query("ticker == 'AMZN'").drop(columns='ticker')\
    .apply(lambda x: x.sub(x.mean()).div(x.std()))
AMZN_z_scores
```
