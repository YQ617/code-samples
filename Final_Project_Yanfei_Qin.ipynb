## 4. Prepare the ML algorithms

### 4.1 Create train and test sets using the selected (by running Random Forest Importance Option) features 

```python
Ab
```

```python
Ab.columns
```

```python
# As I ran Random Forest Important option in Homework 4, I will first continue my study using the 5 most important features only, 
# and if the result is not predicted good enough, I will increase the range of feature selection.
# the reason for initially selecting only the 5 most important features into the model will be explained in the writeup.
Ab_IF = Ab[['Reason for absence','Day of the week','Work load Average/day ','Month of absence','Hit target']]
```

```python
Ab_IF
```

```python
N_y = Ab['Absenteeism time in hours']
N_X = Ab_IF
```

```python
from sklearn.model_selection import train_test_split
N_X_train, N_X_test, N_y_train, N_y_test = train_test_split(
    N_X, N_y, test_size=0.5, random_state=10
)
```

### 4.2 Select multiple algorithms


```python
#failed attempt of trying to use linear regression model, not a model selected but to show that regression models are not working so well with our data
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

#method 1
pipe_fail = make_pipeline(StandardScaler(),
                        PCA(n_components=2),
                        LinearRegression().fit(N_X, N_y))

pipe_fail.fit(N_X_train, N_y_train)
print('Linear Regression (failed) Test Accuracy: %.3f' % pipe_fail.score(N_X_test, N_y_test))
```

```python
#first model: Suuport Vector
from sklearn import svm

pipe1 = make_pipeline(StandardScaler(),                   
                      svm.SVC(decision_function_shape='ovo', kernel='linear', C=9))
```

```python
#second model: Random Forest
from sklearn.ensemble import RandomForestClassifier
pipe2 = make_pipeline(StandardScaler(),                     
                      RandomForestClassifier(n_estimators=150, random_state=50, max_depth=5, criterion='entropy'))
```

```python
#third model: Decision Trees
from sklearn import tree
pipe3 = make_pipeline(StandardScaler(),                     
                      tree.DecisionTreeClassifier(max_depth=3,
                              random_state=50,
                              criterion='entropy'))
```

```python
#fourth model: Stochastic Gradient Descent
from sklearn.linear_model import SGDClassifier
pipe4 = make_pipeline(StandardScaler(),                     
                      SGDClassifier(loss="hinge", penalty="l2", max_iter = 1000, random_state=50))
```

### 4.3 Complete a 10-fold Cross-validation on sets and select the highest performing model.


```python
#model comparison using 10-fold CV
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

kfold = KFold(n_splits=10).split(N_X_train, N_y_train)
scores_SVC = cross_val_score(estimator=pipe1,
                         X=N_X_train,
                         y=N_y_train,
                         cv=kfold,
                         n_jobs=1)
print('SVC CV accuracy scores: %s' % scores_SVC)
print('SVC CV accuracy: %.3f +/- %.3f' % (np.mean(scores_SVC), np.std(scores_SVC)))
```

```python
kfold = KFold(n_splits=10).split(N_X_train, N_y_train)
scores_RF = cross_val_score(estimator=pipe2,
                            X=N_X_train,
                            y=N_y_train,
                            cv=kfold,
                            n_jobs=1)
print('Random Forest CV accuracy scores: %s' % scores_RF)
print('Random Forest CV accuracy: %.3f +/- %.3f' % (np.mean(scores_RF), np.std(scores_RF)))
```

```python
kfold = KFold(n_splits=10).split(N_X_train, N_y_train)
scores_DT = cross_val_score(estimator=pipe3,
                         X=N_X_train,
                         y=N_y_train,
                         cv=kfold,
                         n_jobs=1)
print('Decision Trees CV accuracy scores: %s' % scores_DT)
print('Decision Trees CV accuracy: %.3f +/- %.3f' % (np.mean(scores_DT), np.std(scores_DT)))
```

```python
kfold = KFold(n_splits=10).split(N_X_train, N_y_train)
scores_SGD = cross_val_score(estimator=pipe4,
                         X=N_X_train,
                         y=N_y_train,
                         cv=kfold,
                         n_jobs=1)
print('SGD CV accuracy scores: %s' % scores_SGD)
print('SGD CV accuracy: %.3f +/- %.3f' % (np.mean(scores_SGD), np.std(scores_SGD)))
```
